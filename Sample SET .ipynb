### Data Cleaning

```python
df = data.drop('Unnamed: 0',axis=1)
df
```

```python
df.isnull().sum()
```

```python
df.replace(np.nan,'0.0',inplace = True)
df.isnull().sum()
```

```python
df
```

```python
df.rename({'Cube ID': 'Cube_Id'}, axis=1, inplace=True)
df.rename({'Temperature Layer A': 'Temperature_Layer_A'}, axis=1, inplace=True)
df.rename({'Temperature Layer B': 'Temperature_Layer_B'}, axis=1, inplace=True)
df.rename({'Humidity Layer A': 'Humidity_Layer_A'}, axis=1, inplace=True)
df.rename({'Humidity Layer B': 'Humidity_Layer_B'}, axis=1, inplace=True)
df.head()
```

```python
df['Temperature_Layer_A'] = df.Temperature_Layer_A.str.replace('°C', '').astype(float).round(2)
df['Temperature_Layer_B'] = df.Temperature_Layer_B.str.replace('°C', '').astype(float).round(2)
df['Humidity_Layer_A'] = df.Humidity_Layer_A.str.replace('g/m3', '').astype(float).round(2)
df['Humidity_Layer_B'] = df.Humidity_Layer_B.str.replace('g/m3', '').astype(float).round(2)
df
```

```python
df.describe()
```
