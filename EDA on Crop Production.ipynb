## Univariate Analysis

### Histogram

```python
plt.hist(df['Season'])
```

**Inference:** From this figure we get to know about variety of Crops production in a perticular season. Here variety of Crops in Kharif season are drastically high as compare to others And variety of crops in Summer, Winter and autumn are drastically low. Here Rabi and Whole Year produces average number of various Crops. From which we can conclude that season Kharif is suitable for most of the Crops Summer, Winter and Autumn are suitable for some selective Crops.

### Count Plot

```python
sns.countplot(x=df['Crop_Year'])
plt.xticks(rotation=90)
```

**Inference:** From this figure we get to know about variety of Crops production in a perticular Year. The dataset is from the year 1997 to 2015. Here we have noticed that Variety of Crops have increased during 1997 to 2003 and have decreased afterwards. In the year 2015 there is a certain downfall in the number of various Crops. From which we can conclude that maybe in that year some unexpected incidents or policies were taken. 

### Bivariate Analysis

#### Bar Plot

```python
plt.figure(figsize=(25,15))
sns.barplot(df["State_Name"],df["Production"])
plt.xticks(rotation=90)
```

**Inference:** The figure shows about the Production rate of Crops by every States of India during 1997 to 2015. As per the data Kerala's production rate is drastically high as compare to others. Some states like Bihar, Chattisgarh, Jharkhand, Himachal Pradesh, JK and most of the North Eastern States Produces very less amount of Crops So we can neglect those states from this Crop Production Analysis. From this figure we can also observe that Production rate of most of the costal states are high so maybe because of high production rate of costal Crops their(Costal States) production rate is also high.
